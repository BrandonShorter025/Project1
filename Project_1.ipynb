```python
import pandas as pd

# Create a pivot table to find the average Crude Birth Rate by Year
pivot_table = pd.pivot_table(df, values='Crude Birth Rate', index='Year', aggfunc='mean')

# Display the pivot table
print(pivot_table)
```

```python
import pandas as pd

# Create a pivot table to find the total Birth Number by Year
pivot_table_births = pd.pivot_table(df, values='Birth Number', index='Year', aggfunc='sum')

# Display the pivot table
print(pivot_table_births)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Group by Year and sum the Birth Number
yearly_births = df.groupby('Year')['Birth Number'].sum()

# Create a bar plot for total births by year
plt.figure(figsize=(12,7))
yearly_births.plot(kind='bar')

plt.title('Total Birth Number by Year')
plt.xlabel('Year')
plt.ylabel('Birth Number')

# Show every 5th year (adjust as needed)
plt.xticks(ticks=np.arange(0, len(yearly_births), step=5), rotation=45)

plt.grid(True)
plt.show()
```

This bar graph shows the birth number throughout the years for the USA. This graph illustrates that the birth number fluctuated throughout the years. In the 1920s and 1930s, the birth number was at an all time low. Then its started to raise over the next few decades until it reach another decline period in the 1970s. Finally, the birth number continued to rise from until current times.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(df['Year'], df['General Fertility Rate'], label='General Fertility Rate', color='blue', marker='o')
plt.axvspan(1929, 1939, color='gray', alpha=0.3, label='Great Depression')
plt.axvspan(1946, 1964, color='green', alpha=0.3, label='Baby Boom')
plt.axvspan(1970, 1980, color='orange', alpha=0.3, label='1970s Decline')
plt.title('Fertility Rate Trends Across Historical Periods')
plt.xlabel('Year')
plt.ylabel('General Fertility Rate')
plt.legend()
plt.grid(True)
plt.show()
```

The graph that you see above represents the feetility rates from the 1909 to 2018. As you can see, I included significant historical periods that may have affected the numbers. As you in can see the rates, rose in the 50s and 60 and dropped significantly in the 70s and in recent times.

How have crude birth rates and general fertility rates trended over different decades, and what historical events correlate with significant changes?

The crude birth rates and general fertility rates have generally trended downward from 1909 to 2018. However, there were notable exceptions to this downward trend. From the early 1940s to the early 1960s, both rates saw a sharp increase, largely due to the post-World War II Baby Boom.

After this surge, both birth and fertility rates began to decline, reaching a historic low in the 1970s. This decline can be partially attributed to significant social changes, particularly the landmark Roe v. Wade decision in 1973, which legalized abortion and provided greater access to reproductive healthcare, including birth control.

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(12,6))

# Plot General Fertility Rate on the primary y-axis
color = 'tab:blue'
ax1.set_xlabel('Year')
ax1.set_ylabel('General Fertility Rate', color=color)
ax1.plot(df['Year'], df['General Fertility Rate'], label='General Fertility Rate', color=color, marker='o')
ax1.tick_params(axis='y', labelcolor=color)

# Create a secondary y-axis for the Birth Number
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Birth Number', color=color)
ax2.plot(df['Year'], df['Birth Number'], label='Birth Number', color=color, marker='x')
ax2.tick_params(axis='y', labelcolor=color)

# Highlight historical periods
ax1.axvspan(1929, 1939, color='gray', alpha=0.3, label='Great Depression')
ax1.axvspan(1946, 1964, color='green', alpha=0.3, label='Baby Boom')
ax1.axvspan(1970, 1980, color='orange', alpha=0.3, label='1970s Decline')

# Combine legends from both y-axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2)

plt.title('General Fertility Rate and Birth Number Trends Across Historical Periods')
plt.grid(True)
plt.show()
```

This graph is a combination of both general fertility rates and birth number throughout the ages. As you can see, both of the rates rose and fell at the same rates throught the ages with the exception of the the dfirst couple of years.
